## **1. 이미지 생성하기**

####**&nbsp;&nbsp;0) 기존 네트워크 구조**

: 신경망을 통해 '분류'한다는 것은, 입력 값을 줄여 클래스당 하나의 출력값을 가지게끔 만드는 행위임

: 즉, 데이터를 감소시킴

> cf. '생성'은 데이터를 확장하여 얻어내는 것

<br>

####**&nbsp;&nbsp;1) 백쿼리(backquery)**

######**&nbsp;&nbsp;&nbsp;&nbsp;(1) 개념**

: 기존 네트워크를 뒤집어, input으로 레이블을 주면 신경망을 통해 이미지가 생성되는 것

<br>

######**&nbsp;&nbsp;&nbsp;&nbsp;(2) 백쿼리를 통해 만들어진 이미지의 특성**
: 같은 원핫 인코딩 벡터이면 같은 결과를 출력한다

&nbsp;&nbsp;&nbsp;&nbsp;⇒ [이상(理想)] 각각 서로 다른 이미지 생성

: 그 레이블을 나타내는 모든 훈련 데이터의 뭔가 평균적인 이미지가 나온다.

&nbsp;&nbsp;&nbsp;&nbsp;⇒ [이상(理想)] 평균적인 모호한 이미지가 아닌, 훈련 샘플처럼 보이는 이미지 생성

<br>

*▶ 즉, 간단한 백쿼리로는 목적을 달성하지 못함*

<br>

####**&nbsp;&nbsp;2) 적대적 훈련**

######**&nbsp;&nbsp;&nbsp;&nbsp;(1) 개념**

: 2014년, 이언 굿펠로가 제안

: 생성기(generator)와 판별기(discriminator) 개념의 도입
* 생성기 - 진짜와 같은 그럴듯한 이미지를 생성 (목적 : 판별기가 구분하지 못할 정도의 이미지를 생성하는 것)

* 판별기 - 진짜(1)와 가짜로 생성된 이미지(0)를 구분

<br>

######**&nbsp;&nbsp;&nbsp;&nbsp;(2) 생성기의 훈련 방법**
: 손실함수 활용
* 판별기를 무사히 속이면 생성기에 보상을 줌

* 판별기에 잡히면 생성기에 벌을 줌

: 판별기가 훈련을 거치며 점점 성능이 좋아질수록, 생성기 또한 reward를 통해 훈련이 되어 성능이 증가함

<br>

*▶ 즉, 판별기와 생성기는 서로 적대적 관계로 경쟁을 하게 되며, 이로써 둘 다 성능이 좋아지게 됨*

*▶ 이 구조가 "**생성적 적대 신경망**(GAN, Generative Adversarial Network)"*

<br>
<br>
